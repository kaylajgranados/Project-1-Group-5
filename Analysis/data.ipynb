```python
import pandas as pd
```

```python
new_city_revenue_df = city_revenue_df[["Entity Name", "Fiscal Year", 
                            "Total Revenues", "Estimated Population"]]
new_city_revenue_df.head()
```

```python
# LA vs Thousand Oaks- 2020 report.

# Pull revenue of LA and Thousand Oaks from new_city_revenue_df
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8' )

# Locate city rows in 2020
los_angeles = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="Los Angeles") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]
thousand_oaks = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="Thousand Oaks") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]

#get Revenues as a float
la_revenue = los_angeles['Total Revenues'].astype(float).item()
to_revenue = thousand_oaks['Total Revenues'].astype(float).item()

#Convert to currency string for printing (optional)
str_la_rev = locale.currency(la_revenue,grouping=True)
str_to_rev = locale.currency(to_revenue,grouping=True)

#Print Analysis (Optional)
print(f"Comparison for LA County:")
print(f"---------------------------------------------")
print(f"Los Angeles Total Revenue: {str_la_rev}")
print(f"Thousand Oaks Total Revenue: {str_to_rev}")

#Compare revenue sizes
if la_revenue > to_revenue:
    print("Los Angeles generates more total revenue for the city than Thousand Oaks.")
elif to_revenue<la_revenue:
    print("Thousand Oaks generates more total revenue for the city than Los Angeles.")
else:
    print("Los Angeles and Thousand Oaks generate the same amount of mont for the city.")   
```

```python
# San Jose vs Los Gatos- 2020 report.

# Locate city rows in 2020
san_jose = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="San Jose") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]
los_gatos = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="Los Gatos") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]

#get Revenues as a float
sj_revenue = san_jose['Total Revenues'].astype(float).item()
lg_revenue = los_gatos['Total Revenues'].astype(float).item()

#Convert to currency string for printing (optional)
str_sj_rev = locale.currency(sj_revenue,grouping=True)
str_lg_rev = locale.currency(lg_revenue,grouping=True)

#Print Analysis (Optional)
print(f"Comparison for Santa Clara County:")
print(f"---------------------------------------------")
print(f"San Jose Total Revenue: {str_sj_rev}")
print(f"Los Gatos Total Revenue: {str_lg_rev}")

#Compare revenue sizes
if sj_revenue > to_revenue:
    print("San Jose generates more total revenue for the city than Los Gatos.")
elif sj_revenue<lg_revenue:
    print("Los Gatos generates more total revenue for the city than San Jose.")
else:
    print("San Jose and Los Gatos generate the same amount of mont for the city.")    
```

```python
# San Diego vs Poway- 2020 report.

# Locate city rows in 2020
san_diego = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="San Diego") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]
poway = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="Poway") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]

#get Revenues as a float
sd_revenue = san_diego['Total Revenues'].astype(float).item()
po_revenue = poway['Total Revenues'].astype(float).item()

#Convert to currency string for printing (optional)
str_sd_rev = locale.currency(sd_revenue,grouping=True)
str_po_rev = locale.currency(po_revenue,grouping=True)

#Print Analysis (Optional)
print(f"Comparison for San Diego County:")
print(f"---------------------------------------------")
print(f"San Diego Total Revenue: {str_sd_rev}")
print(f"Poway Total Revenue: {str_po_rev}")

#Compare revenue sizes
if sd_revenue > po_revenue:
    print("San Diego generates more total revenue for the city than Poway.")
elif po_revenue<sd_revenue:
    print("Poway generates more total revenue for the city than San Diego.")
else:
    print("San Diego and Poway generate the same amount of mont for the city.")    
      
```

```python
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
import seaborn as sns
from scipy.stats import linregress
from matplotlib.pyplot import figure
from sklearn import datasets
```

```python
# create datafram shows total revenue of LA county
LA_county = {
    'City' : ['Los Angeles' , 'Thousand Oaks'],
    'Total Revenue' : [la_revenue, to_revenue]
}
LA_county_df = pd.DataFrame(LA_county)
LA_county_df
```

```python
# creat chart of Total revenue for LA county
revenue_la_county = [la_revenue,to_revenue]
x_axis = np.arange(len(revenue_la_county))
fig1, ax1 = plt.subplots(figsize=(5, 6))
plt.bar(x_axis, revenue_la_county, color='b', alpha=0.5, align="center")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Los Angeles", "Thousand Oaks"])

plt.title("Total Revenue of LA vs. Thousand Oaks")
plt.xlabel("Cities")
plt.ylabel("Total Revenue")


plt.tight_layout()
plt.show()
```
