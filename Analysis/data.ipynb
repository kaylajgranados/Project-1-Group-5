```python
property_tax_file = "../Resources/Property_tax_etc.csv"
```

```python
property_tax_df = pd.read_csv(property_tax_file)
property_tax_df.head()
```

```python
new_property_tax_df = property_tax_df[["Entity Name", "Fiscal Year", "Secured and Unsecured Property Taxes_Total Revenues_Taxes",
                        "Sales and Use Taxes_Total Revenues_Taxes","Franchises_Total Revenues_Taxes", "Business License Taxes_Total Revenues_Taxes", 
                        "Utility Users Taxes_Total Revenues_Taxes"]]
new_property_tax_df.head()
```

```python
new_property_tax_df = property_tax_df[["Entity Name", "Fiscal Year", "Secured and Unsecured Property Taxes_Total Revenues_Taxes",
                        "Sales and Use Taxes_Total Revenues_Taxes","Transportation Taxes – Transit_Total Revenues_Taxes", "Franchises_Total Revenues_Taxes", "Business License Taxes_Total Revenues_Taxes", 
                        "Utility Users Taxes_Total Revenues_Taxes", "Construction Development Taxes_Total Revenues_Taxes", "Employers Payroll Tax_Total Revenues_Taxes – Other", "Parking Tax_Total Revenues_Taxes – Other"]]
new_property_tax_df.head()
```

```python
# Create a dataframe to compare Property Tax 
only_property_tax_df = new_property_tax_df[["Entity Name", "Fiscal Year", "Secured and Unsecured Property Taxes_Total Revenues_Taxes"]]

# Rename columns 
renamed_property_tax_df = only_property_tax_df.rename(columns={"Secured and Unsecured Property Taxes_Total Revenues_Taxes": "Total Property Tax"})

# Use loc to only display data for our 6 cities for 2020 
los_angeles_property_tax= renamed_property_tax_df.loc[(renamed_property_tax_df['Entity Name']=="Los Angeles") & (renamed_property_tax_df['Fiscal Year']==2020)]
san_jose_property_tax= renamed_property_tax_df.loc[(renamed_property_tax_df['Entity Name']=="San Jose") & (renamed_property_tax_df['Fiscal Year']==2020)]
san_diego_property_tax= renamed_property_tax_df.loc[(renamed_property_tax_df['Entity Name']=="San Diego") & (renamed_property_tax_df['Fiscal Year']==2020)]
thousand_oaks_property_tax= renamed_property_tax_df.loc[(renamed_property_tax_df['Entity Name']=="Thousand Oaks") & (renamed_property_tax_df['Fiscal Year']==2020)]
los_gatos_property_tax= renamed_property_tax_df.loc[(renamed_property_tax_df['Entity Name']=="Los Gatos") & (renamed_property_tax_df['Fiscal Year']==2020),]
poway_property_tax= renamed_property_tax_df.loc[(renamed_property_tax_df['Entity Name']=="Poway") & (renamed_property_tax_df['Fiscal Year']==2020)]

frames = [los_angeles_property_tax, san_jose_property_tax, san_diego_property_tax, thousand_oaks_property_tax, los_gatos_property_tax, poway_property_tax]
property_tax_by_city_df = pd.concat(frames)

# Sort highest to lowest 
highest_property_tax_df = property_tax_by_city_df.sort_values("Total Property Tax", ascending=False)
highest_property_tax_df
```

```python
# Create a dataframe to compare business related taxes 
business_tax_df = new_property_tax_df[["Entity Name", "Fiscal Year", "Sales and Use Taxes_Total Revenues_Taxes", "Franchises_Total Revenues_Taxes", "Business License Taxes_Total Revenues_Taxes"]]

# Rename columns 
renamed_business_tax_df= business_tax_df.rename(columns={"Sales and Use Taxes_Total Revenues_Taxes": "Total Sales Tax", "Franchises_Total Revenues_Taxes": "Total Franchise Tax", "Business License Taxes_Total Revenues_Taxes": "Total Business License Tax"})

# Make overall total column 
cols = ["Total Sales Tax", "Total Franchise Tax", "Total Business License Tax"]
renamed_business_tax_df["Overall Total Business Tax"] = renamed_business_tax_df[cols].sum(axis=1)

# Use loc to only display data for our 6 cities for 2020 
los_angeles_business_tax= renamed_business_tax_df.loc[(renamed_business_tax_df['Entity Name']=="Los Angeles") & (renamed_business_tax_df['Fiscal Year']==2020)]
san_jose_business_tax= renamed_business_tax_df.loc[(renamed_business_tax_df['Entity Name']=="San Jose") & (renamed_business_tax_df['Fiscal Year']==2020)]
san_diego_business_tax= renamed_business_tax_df.loc[(renamed_business_tax_df['Entity Name']=="San Diego") & (renamed_business_tax_df['Fiscal Year']==2020)]
thousand_oaks_business_tax= renamed_business_tax_df.loc[(renamed_business_tax_df['Entity Name']=="Thousand Oaks") & (renamed_business_tax_df['Fiscal Year']==2020)]
los_gatos_business_tax= renamed_business_tax_df.loc[(renamed_business_tax_df['Entity Name']=="Los Gatos") & (renamed_business_tax_df['Fiscal Year']==2020),]
poway_business_tax= renamed_business_tax_df.loc[(renamed_business_tax_df['Entity Name']=="Poway") & (renamed_business_tax_df['Fiscal Year']==2020)]

frames = [los_angeles_business_tax, san_jose_business_tax, san_diego_business_tax, thousand_oaks_business_tax, los_gatos_business_tax, poway_business_tax]
business_tax_by_city_df = pd.concat(frames)

# Sort highest to lowest 
highest_business_tax_df = business_tax_by_city_df.sort_values("Overall Total Business Tax", ascending=False)
highest_business_tax_df
```

```python
#Create a dataframe to compare "Misc" taxes categories (utility, transportation, construction, payroll, etc.)
misc_tax_df = new_property_tax_df[["Entity Name", "Fiscal Year", "Transportation Taxes – Transit_Total Revenues_Taxes",  
                        "Utility Users Taxes_Total Revenues_Taxes", "Construction Development Taxes_Total Revenues_Taxes", "Employers Payroll Tax_Total Revenues_Taxes – Other", "Parking Tax_Total Revenues_Taxes – Other"]]

#Rename Columns 
renamed_misc_tax_df= misc_tax_df.rename(columns={"Transportation Taxes – Transit_Total Revenues_Taxes": "Total Transportation Tax", "Utility Users Taxes_Total Revenues_Taxes": "Total Utility Tax", "Construction Development Taxes_Total Revenues_Taxes": "Total Construction Tax", "Employers Payroll Tax_Total Revenues_Taxes – Other": "Total Payroll Tax", "Parking Tax_Total Revenues_Taxes – Other": "Total Parking Tax"})

# Make Overall Total Column 
cols = ["Total Transportation Tax", "Total Utility Tax", "Total Construction Tax", "Total Payroll Tax", "Total Parking Tax"]
renamed_misc_tax_df["Overall Total Miscellaneous Taxes"] = renamed_misc_tax_df[cols].sum(axis=1)

# Use loc to only display data for our 6 cities for 2020 
los_angeles_misc_tax= renamed_misc_tax_df.loc[(renamed_misc_tax_df['Entity Name']=="Los Angeles") & (renamed_misc_tax_df['Fiscal Year']==2020)]
san_jose_misc_tax= renamed_misc_tax_df.loc[(renamed_misc_tax_df['Entity Name']=="San Jose") & (renamed_misc_tax_df['Fiscal Year']==2020)]
san_diego_misc_tax= renamed_misc_tax_df.loc[(renamed_misc_tax_df['Entity Name']=="San Diego") & (renamed_misc_tax_df['Fiscal Year']==2020)]
thousand_oaks_misc_tax= renamed_misc_tax_df.loc[(renamed_misc_tax_df['Entity Name']=="Thousand Oaks") & (renamed_misc_tax_df['Fiscal Year']==2020)]
los_gatos_misc_tax= renamed_misc_tax_df.loc[(renamed_misc_tax_df['Entity Name']=="Los Gatos") & (renamed_misc_tax_df['Fiscal Year']==2020),]
poway_misc_tax= renamed_misc_tax_df.loc[(renamed_misc_tax_df['Entity Name']=="Poway") & (renamed_misc_tax_df['Fiscal Year']==2020)]
frames = [los_angeles_misc_tax, san_jose_misc_tax, san_diego_misc_tax, thousand_oaks_misc_tax, los_gatos_misc_tax, poway_misc_tax]
misc_tax_by_city_df = pd.concat(frames)

# Sort highest to lowest 
highest_misc_tax_df = misc_tax_by_city_df.sort_values("Overall Total Miscellaneous Taxes", ascending=False)
highest_misc_tax_df
```

```python
# Create a dataframe that only displays the total columns 

entity_name=new_property_tax_df["Entity Name"]
fiscal_year=new_property_tax_df["Fiscal Year"]
total_property = highest_property_tax_df["Total Property Tax"]
total_business = highest_business_tax_df["Overall Total Business Tax"]
total_misc = highest_misc_tax_df["Overall Total Miscellaneous Taxes"]

total_category_by_city_df = pd.DataFrame({"Entity Name": entity_name,
                                "Fiscal Year": fiscal_year, 
                                "Total Property": total_property,
                               "Total Business": total_business, 
                               "Total Misc": total_misc})

#Add columns 
cols = ["Total Property", "Total Business", "Total Misc"] 
total_category_by_city_df["Overall Total Taxes"] = total_category_by_city_df[cols].sum(axis=1)
cols1 = ["Total Property", "Total Business", "Total Misc"] 
total_category_by_city_df["Mean"] = total_category_by_city_df[cols1].mean(axis=1)
cols2 = ["Total Property", "Total Business", "Total Misc"] 
total_category_by_city_df["Median"] = total_category_by_city_df[cols1].median(axis=1)
```

```python
# Use loc to only display data for our 6 cities for 2020 -this is broken 
los_angeles_total_tax= total_category_by_city_df.loc[(total_category_by_city_df['Entity Name']=="Los Angeles") & (total_category_by_city_df['Fiscal Year']==2020)]
san_jose_total_tax= total_category_by_city_df.loc[(total_category_by_city_df['Entity Name']=="San Jose") & (total_category_by_city_df['Fiscal Year']==2020)]
san_diego_total_tax= total_category_by_city_df.loc[(total_category_by_city_df['Entity Name']=="San Diego") & (total_category_by_city_df['Fiscal Year']==2020)]
thousand_oaks_total_tax= total_category_by_city_df.loc[(total_category_by_city_df['Entity Name']=="Thousand Oaks") & (total_category_by_city_df['Fiscal Year']==2020)]
los_gatos_total_tax= total_category_by_city_df.loc[(total_category_by_city_df['Entity Name']=="Los Gatos") & (total_category_by_city_df['Fiscal Year']==2020),]
poway_total_tax= total_category_by_city_df.loc[(total_category_by_city_df['Entity Name']=="Poway") & (total_category_by_city_df['Fiscal Year']==2020)]

frames = [los_angeles_total_tax, san_jose_total_tax, san_diego_total_tax, thousand_oaks_total_tax, los_gatos_total_tax, poway_total_tax]
total_tax_by_city_df = pd.concat(frames)
```
