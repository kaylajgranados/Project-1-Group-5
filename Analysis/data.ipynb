```python
# LA vs Thousand Oaks- 2020 report.

# Pull revenue of LA and Thousand Oaks from new_city_revenue_df
import locale
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8' )

# Locate city rows in 2020
los_angeles = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="Los Angeles") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]
thousand_oaks = new_city_revenue_df.loc[(new_city_revenue_df['Entity Name']=="Thousand Oaks") & (new_city_revenue_df['Fiscal Year']==2020),['Total Revenues']]

#get Revenues as a float
la_revenue = los_angeles['Total Revenues'].astype(float).item()
to_revenue = thousand_oaks['Total Revenues'].astype(float).item()

#Convert to currency string for printing (optional)
str_la_rev = locale.currency(la_revenue,grouping=True)
str_to_rev = locale.currency(to_revenue,grouping=True)

#Print Analysis (Optional)
print(f"Comparison for LA County:")
print(f"---------------------------------------------")
print(f"Los Angeles Total Revenue: {str_la_rev}")
print(f"Thousand Oaks Total Revenue: {str_to_rev}")

#Compare revenue sizes
if la_revenue > to_revenue:
    print("Los Angeles generates more total revenue for the city than Thousand Oaks.")
elif to_revenue<la_revenue:
    print("Thousand Oaks generates more total revenue for the city than Los Angeles.")
else:
    print("Los Angeles and Thousand Oaks generate the same amount of mont for the city.")   
```

```python
# Create new DF with Density and Total Revenues, set index to the Entity Name

rev2020_df = city_revenue_df.loc[city_revenue_df['Fiscal Year']==2020]
density_rev_df = pd.merge(new_density_df,rev2020_df,on='Entity Name',how='right').dropna(how='any')
density_rev_df.set_index('Entity Name')

#pulling series of revenue, density, and total population
density = density_rev_df['density']
revenue = density_rev_df['Total Revenues']
totalPopulation = density_rev_df['Estimated Population']
revPerCapita = density_rev_df['Revenues Per Capita']

#function to check for correlation
def findCorrelation(series1,series2):
    correlation = st.pearsonr(series1,series2)
    print(f"The correlation between these two factors is {round(correlation[0],2)}")
    if correlation[0] < .3:
        print("There is little to no evidence of a correlation between these 2 factors.")
    elif (correlation[0] > .3) & (correlation[0] <= .5):
        print("There may be a weak correlation between these 2 factors.")
    elif (correlation[0] > .5) & (correlation[0] <= .7):
        print("The data suggests a weak correlation between these 2 factors.")    
    elif (correlation[0] > .7) & (correlation[0] <1):    
        print("There is a strong likelihood that these two factors are correlated.")
    elif correlation[0] ==1:
        print("There is a definite correlation between these two factors.") 


#check correlation of Revenue and Population
print(f"Finding the correlation between Total Population and Revenue:")
findCorrelation(totalPopulation,revenue)  
print(f"--------------------------------------------------------------")   

#check correlation of Revenue and Density
print(f"Finding the correlation between Density and Revenue:")
findCorrelation(density,revenue)  
print(f"--------------------------------------------------------------") 

#check correlation of Revenue and Density
print(f"Finding the correlation between Density and Revenue per Capita:")
findCorrelation(revPerCapita,density)  
print(f"--------------------------------------------------------------")
# Possible Conclusion, the data suggests that density does not indicate more wealth for the city. :(



#check the p-values for significance
pop_p_val = st.normaltest(totalPopulation.sample(50))
den_p_val = st.normaltest(density.sample(50))   
perCap_p_val = st.normaltest(revPerCapita.sample(50))   
```

```python
# for Total Population and Revenues
x_values = density_rev_df['Estimated Population']
y_values = density_rev_df['Total Revenues']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)
```

```python
# create plot for Polulation and Revenue
fig1, ax1 = plt.subplots(figsize=(5, 6))
plt.scatter(x_values,y_values,s=50, color="blue")
plt.plot(x_values,regress_values,"r-")
plt.title('Regression Plot of Total Population and Revenues')
plt.xlabel('Population')
plt.ylabel('Revenues')
plt.annotate(line_eq, (0.6*1e6,0.5*1*9),fontsize=10,color="red")

print(f"The r-squared is: {rvalue**2}")

plt.show()
```
